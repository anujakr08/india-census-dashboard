```python
import plotly.express as px
import pandas as pd


# Function to plot the map based on the user-selected state
def plot_state_on_map(state_name):
    # Filter the DataFrame for the selected state
    selected_state_df = state_df[state_df['State'] == state_name]
    
    # Create the scatter mapbox plot
    fig = px.scatter_mapbox(selected_state_df, 
                            lat="Latitude", 
                            lon="Longitude", 
                            hover_name="State", 
                            zoom=4, 
                            size_max=50, 
                            mapbox_style="carto-positron", 
                            width=1200, 
                            height=700)
    
    # Set the map's center to the state's coordinates
    fig.update_layout(mapbox_center={"lat": selected_state_df.iloc[0]['Latitude'], 
                                     "lon": selected_state_df.iloc[0]['Longitude']}, 
                      mapbox_zoom=5)
    
    fig.show()

# Example usage
# state = input("Enter a state name: ")  # User input
plot_state_on_map('Punjab')
```

## top 5 State with  high category

```python
import plotly.express as px

def plot_literacy_rate(category_col, rate_col, category_name):
    # Group by 'State' and calculate total population and the specified literate population
    pop = final_df.groupby('State')['Population'].sum().reset_index()
    literate = final_df.groupby('State')[category_col].sum().reset_index()
    
    # Merge population and literate data, calculate the literacy rate
    merged_data = pop.merge(literate, on='State')
    merged_data[rate_col] = round((merged_data[category_col] / merged_data['Population']) * 100)
    
    # Merge with state_df to get Latitude and Longitude
    merged_data = merged_data.merge(state_df, on='State')
    
    # Get the top 5 states based on the calculated literacy rate
    top5 = merged_data.sort_values(by=rate_col, ascending=False).head(5)
    
    # Create a scatter mapbox plot
    fig = px.scatter_mapbox(
        top5, 
        lat="Latitude", 
        lon="Longitude", 
        color_continuous_scale='Viridis',  # Color scale for visualization
        size_max=25, 
        zoom=5, 
        mapbox_style="carto-positron",  # Map style
        width=1200, 
        height=700, 
        hover_name='State',  # Hover will display the State name
        color=rate_col,  # Color the points by the literacy rate
        hover_data={'State': True, 'Latitude': False, 'Longitude': False}  # Show only 'State' in hover
    )
    
    # Display the figure
    fig.show()

def plot_population():
    # Group by 'State' and calculate total population
    pop = final_df.groupby('State')['Population'].sum().reset_index()
    
    # Merge with state_df to get Latitude and Longitude
    merged_data = pop.merge(state_df, on='State')
    
    # Get the top 5 states based on population
    top5 = merged_data.sort_values(by='Population', ascending=False).head(5)
    
    # Create a scatter mapbox plot
    fig = px.scatter_mapbox(
        top5, 
        lat="Latitude", 
        lon="Longitude", 
        color_continuous_scale='Viridis',  # Color scale for visualization
        size_max=25, 
        zoom=5, 
        mapbox_style="carto-positron",  # Map style
        width=1200, 
        height=700, 
        hover_name='State',  # Hover will display the State name
        color='Population',  # Color the points by population
        hover_data={'State': True, 'Latitude': False, 'Longitude': False}  # Show only 'State' in hover
    )
    
    # Display the figure
    fig.show()

def state_category(category):
    if category == 'Literacy Rate':
        plot_literacy_rate('Literate', 'State_literacy', 'Literacy Rate')
    elif category == 'Male Literacy Rate':
        plot_literacy_rate('Male_Literate', 'Male_literacy', 'Male Literacy Rate')
    elif category == 'Female Literacy Rate':
        plot_literacy_rate('Female_Literate', 'Female_literacy', 'Female Literacy Rate')
    elif category == 'Population':
        plot_population()

# Example usage
state_category('Population')
```
